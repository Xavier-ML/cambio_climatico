```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
emissions = pd.read_csv('Resources/co2_emission.csv')
emissions.columns = ['Entity','Code','Year','Annual CO2 Emissions (tonnes)']
# emissions['Year'] = pd.to_datetime(emissions['Year'])
emissions.head(0)
```

```python
plt.figure(figsize = (18, 10))

sns.lineplot(data = emissions, 
             x = 'Year', 
             y = 'Annual CO2 Emissions (tonnes)', 
             linewidth = 1, 
             hue = 'Entity', 
             alpha = .8)
plt.grid(alpha = .4)
plt.legend([],[], frameon=False)
plt.show()
```

```python
'''
Can see above that emission levels have sky-rocketed since ~1950. I will compare this to global temperature data.
With climate data it is important to remember that bodies of water absorb significant amounts of energy, distorting
air teperature data.'''

land_temp = pd.read_csv('/Users/christianstankovic/Desktop/CLIMATE_CHANGE/Resources/Earth_Surface_Data/GlobalTemperatures.csv')
land_temp['dt'] = pd.to_datetime(land_temp['dt'])
land_temp['year'] = land_temp['dt'].dt.year
land_temp.head(0)
```

```python
yearly_average = land_temp.groupby('year').mean()
```

```python
plt.figure(figsize = (18, 14))

ax = sns.lineplot(data = yearly_average,
            x = yearly_average.index,
            y = 'LandAverageTemperature',
            linewidth = .5,
            alpha = .8,
            color = 'maroon',
            legend = False)

ax1 = ax.twinx()

sns.lineplot(data = emissions, 
             x = 'Year', 
             y = 'Annual CO2 Emissions (tonnes)', 
             linewidth = 1, 
             hue = 'Entity', 
             alpha = .8,
             legend = False,
             ax = ax1)




# plt.grid(alpha = .4)
# plt.legend([],[], frameon=False)
plt.show()
```

```python
'''
the average global temperatures have begun following the trend of carbon emissions, starting in the early 1900s.
It can be seen that the climb in average temperature has exceeded the 1.5 C climb that scientists have
warned about, and is reaching the 3 C that would result in catastrophic damage.'''
```
